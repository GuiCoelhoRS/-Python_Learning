### Python Variables

##### Rules for Variables
- Must start with a letter or the underscore
- It cannot start with a number
- Can only contain alpha numeric-numbers and underscores ( a-z , A-Z , 0-9 and _ ) 
- Are Case Sensitive
- It cannot be any of the Python Keywords

```python
# Python variables don't need to be specified but they can
x = 32        # Type Int
y = "Ana"     # Type String -> Can be "" or ''
z = 3.2       # Type Float

# Variables with specification
a = int(3)   # 3
b = str(3)   # "3"
c = float(3) # 3.0

# Getting the type of a variable can be done with -> type()
print(type(x))
print(type(y))
print(type(z))

# Assigning multiple variables
## Many Values to Multiple Variables
r,g,f = "Orange" , "Banana" , "Apple"
## One Value to Multiple Variables
u = i = o = "Orange"

# Variable Output using print()
print(x)
print(c)
print(x , y)
## We can print calculations if we use two numbers like this
print(x + z) # 32 + 3.2 

# Global variables - Can be acceced everywhere in the document
global num ;
num = 2



# -> Variables are not case sensitive a is different than A
```

### Strings


```python
# Normal Strings
print("Hello")  # Double Quotes
print('Hello')  # Single Quotes

# Strings with Quotes Inside 
print("H'e'llo")
print('H"e"llo')

# Multiline Strings
a = """Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua."""
print(a) 
```

##### Strings are arrays and we can do a lot because of that


```python
# Acessing Elements of a String
a = ("Hello World!")
print(a[1]) # [pos in the String]
print("---")

# Looping trough a String
for x in "Hi" :
  print(x)

print("---")

# Calculating the Length of a String
print(len(a)) # len(String)
print("---")

# Check if a certain phrase or character is inside the String
print("world" in a)
print("World" in a)
# or 
if "World" in a:
  print("World is in the String a")
print("---")

# Check IF NOT
# the same thing but `not in`


### -------------------------------------------------------------------------------------------------
# Slicing Strings
b = "Hello World!"
print(b[2:5])
print(b[:5])
print(b[5:])
    # Negative Indexes Start the slice from the end of the String
print(b[-2:])
print("---")

# String Modifiyng
c = " Hello, World!"
print(c.upper())
print(c.lower())
print(c.strip()) # Removes white spaces from the start and end
print(c.replace("H","J")) #  Replaces a String with another String
print(c.split(",")) # Splits the string if it finds the separator in this case ,
print("---")

# String Concatenation
d = "Hello"
e = "World"
f = d + e
k = d + "   a    " + e
print(f)
print(k)
print("---")

# Format Strings
## This is the way we can combine strings and numbers
age = 35 
txt = f"My name is John, and I am  {age} years old" # Notice the f before the string thats the format part
print(txt)
print("---")

# Adding Ilegall characters in Strings
txt2 = "We are the so-called \"Vikings\" from the north." # Here with the \ we can use double quotes that normally we couldn't use
print(txt2)
  ## \'  - Single Quote
  ## \\  - Backslash
  ## \n  - New Line
  ## \r  - Carriage Return
  ## \t  - Tab
  ## \b  - Backspace
  ## \ooo - Octal Value
  ## \xhh - Hex Value
```

#### Python String Methods Cheatsheet

Case Conversion

    capitalize(): Converts the first character to upper case.

    title(): Converts the first character of each word to upper case.

    upper(): Converts string into upper case.

    lower(): Converts string into lower case.

    casefold(): Converts string into lower case (more aggressive than lower()).

    swapcase(): Swaps cases (lower becomes upper and vice versa).

Checks & Validation (Returns True/False)

    isalnum() / isalpha(): Checks if characters are alphanumeric / alphabetic.

    isdecimal() / isdigit() / isnumeric(): Checks if characters are decimals / digits / numeric.

    isascii() / isidentifier(): Checks if characters are ASCII / valid identifier.

    islower() / isupper() / istitle(): Checks the casing of the string.

    isprintable() / isspace(): Checks if printable / whitespace.

    startswith() / endswith(): Checks if string starts/ends with a specified value.

Search & Indexing

    count(): Returns the number of times a value occurs in a string.

    find() / rfind(): Searches for a value and returns the position (or -1 if not found).

    index() / rindex(): Searches for a value and returns the position (raises error if not found).

    Note: r prefix indicates searching from the Right (end).

Formatting & Alignment

    center(): Returns a centered string.

    ljust() / rjust(): Returns a left / right justified version of the string.

    zfill(): Fills the string with 0 values at the beginning.

    format() / format_map(): Formats specified values in a string.

    expandtabs(): Sets the tab size of the string.

Modification (Split, Join, Strip)

    join(): Joins elements of an iterable to the end of the string.

    split() / rsplit(): Splits the string at a separator into a list.

    splitlines(): Splits the string at line breaks.

    partition() / rpartition(): Splits string into a tuple of three parts.

    strip() / lstrip() / rstrip(): Trims whitespace from both ends / left / right.

    replace(): Replaces a specified value with another value.

Encoding & Translation

    encode(): Returns an encoded version of the string.

    maketrans(): Creates a translation table.

    translate(): Returns a translated string using a translation table.
