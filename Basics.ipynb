### Python Variables

##### Rules for Variables
- Must start with a letter or the underscore
- It cannot start with a number
- Can only contain alpha numeric-numbers and underscores ( a-z , A-Z , 0-9 and _ ) 
- Are Case Sensitive
- It cannot be any of the Python Keywords

```python
# Python variables don't need to be specified but they can
x = 32        # Type Int
y = "Ana"     # Type String -> Can be "" or ''
z = 3.2       # Type Float

# Variables with specification
a = int(3)   # 3
b = str(3)   # "3"
c = float(3) # 3.0

# Getting the type of a variable can be done with -> type()
print(type(x))
print(type(y))
print(type(z))

# Assigning multiple variables
## Many Values to Multiple Variables
r,g,f = "Orange" , "Banana" , "Apple"
## One Value to Multiple Variables
u,i,o = "Orange"

# -> Variables are not case sensitive a is different than A
```
